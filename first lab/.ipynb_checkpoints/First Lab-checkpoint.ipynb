```python
d1 = {'Name': ['Pankaj', 'Meghna', 'Lisa', 'Ahmad'], 'ID': [1, 2, 3, 4], 'Country': ['India', 'India', 'USA', 'Palestine'],
      'Role': ['CEO', 'CTO', 'CTO', 'COO']}
df1 = pd.DataFrame(d1)

df2 = pd.DataFrame({'ID': [1, 2, 3, 7], 'Name': ['Pankaj', 'Anupam', 'Amit', 'Mazen']})
```

```python
pd.merge(df1, df2)
```

```python
# Inner Join:
df = pd.merge(df1, df2, on='ID')
df
```

```python
# Left Outer Join:
df_left = df1.merge(df2, how='left', on='ID')
df_left
```

```python
# Full Outer Join:
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

```python
# Left Anti-Join:
df3 = df1.merge(df2, on='ID', how='left', indicator=True)
df_left_anti = df3.loc[df3['_merge'] == 'left_only', 'ID']
d = df1[df1['ID'].isin(df_left_anti)]
d
```

```python
# Right Outer Join:
df_right = pd.merge(df1, df2, how='right')
df_right
```

```python
# Right Anti-Join:
df3 = df1.merge(df2, on='ID', how='right', indicator=True)
df = df3.loc[df3['_merge'] == 'right_only', 'ID']
d = df2[df2['ID'].isin(df)]
d
```

```python
# Full Anti-Join:
df3 = df1.merge(df2, on='ID', how='outer', indicator=True)
df_full_anti = df3.loc[df3['_merge'] != 'both', 'ID']
d = df3[df3['ID'].isin(df_full_anti)]
d
```

```python
np.random.seed(54)

n_rows = 1000000
n_columns = 100

data = np.random.randn(n_rows, n_columns)  # Random data
columns = [f'Column_{i}' for i in range(1, n_columns+1)]  # Column names

dataFrame = pd.DataFrame(data, columns=columns)
```

```python
%time
kurt = dataFrame.kurt(axis=1);
kurt
```

$$ 
\text{kurt} = \frac{n(n+1)}{(n-1)(n-2)(n-3)} \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{s} \right)^4 - \frac{3(n-1)^2}{(n-2)(n-3)}
$$

- $n$ size.
- $\bar{x}$ is the sample mean.
- $s$ is standard deviation.



```python
%time

n = dataFrame.shape[1]
m = dataFrame.mean(axis=1)
std = dataFrame.std(axis=1)

sum_term = ((dataFrame.sub(m, axis=0).div(std, axis=0))**4).sum(axis=1)
kurtosis_values = (n * (n+1) / ((n-1) * (n-2) * (n-3)) * sum_term) - 3 * (n-1)**2 / ((n-2) * (n-3))

kurtosis_values
```
