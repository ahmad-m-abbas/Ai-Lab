```python
import pandas as pd
```

```python
dinner = pd.read_csv('Dinner.csv')
lunch = pd.read_csv('Lunch.csv')
waiters = pd.read_csv('Waiters.csv')
```

```python
# 1.
all = pd.merge(dinner, lunch, how='outer')
all
```

```python
# 1.
all_with_type = pd.merge(waiters[waiters['type'] == 'part time'], all, how='left')
all_part_time = all_with_type[all_with_type['type'] == 'part time']
all_part_time.groupby('Waiter').sum()[['tip']].sort_values(['tip'], ascending=False)
```

```python
# 2 + 3.
all_types = pd.merge(waiters, all, how='left')
a = all_types.groupby('Waiter', sort=True).sum()
ans2 = a[['total_bill', 'tip']]
ans2
```

```python
# 3.
all_types['count'] = all_types['size'] > 3
ans3 = all_types.groupby('Waiter').sum()['count']
ans3
```
